#Customer Churn Analysis#An Exploratory Data Analysis using Python#

```python

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading Customer Attrition data and storing into df1

df1 = pd.read_csv("Customer_Data/Customer Attrition Status.csv")
df1.head()
```

```python
## Reading Customer Demographics data and storing into df2

df2 = pd.read_csv("Customer_Data/Customer Demographics.csv")
df2.head()
```

```python
# Reading Customer Investment data and storing into df3

df3= pd.read_csv("Customer_Data/Customer Investment Snapshot.csv")
df3.head()
```

```python
# Reading Customer Portfolio data and storing into df4

df4= pd.read_csv("Customer_Data/Customer Portfolio Snapshot.csv")
df4.head()
```

```python
# Joining(Inner) Customer Attrition ad demographics data based on rownumber and Customer_id 

df5=pd.merge(df1,df2,on=['CustomerId','RowNumber'])
df5.head()
```

```python
# Joining(Inner) df5 and df3 data based on rownumber and Customer_id 

df6=pd.merge(df5,df3,on=['CustomerId','RowNumber'])
df6.head()
```

```python
#Joining(Inner) Customer Attrition ad demographics data based on rownumber and Customer_id 

df7=pd.merge(df6,df4,on=['CustomerId','RowNumber'])
df7.head()
```

```python
# checking datatypes of column

df7.dtypes
```

```python
#coverting datatype of column as per requirements

df7["CustomerId"]=df7["CustomerId"].astype(int)
df7["Exited"]=df7["Exited"].astype(int)
df7["Age"]=df7["Age"].astype(int)
df7["Tenure"]=df7["Tenure"].astype(int)
#df7["CreditScore"]=df7["CreditScore"].astype(int) ( will change in later steps)
df7["NumOfProducts"]=df7["NumOfProducts"].astype(int)
df7["HasChckng"]=df7["HasChckng"].astype(int)
df7["IsActiveMember"]=df7["IsActiveMember"].astype(int)
```

```python
# verifying the datatypes of column after changing the datatype

df7.dtypes
```

```python
#drops the row/columns if all values are null.

df7.dropna(how = 'all')
```

```python
df7.isnull().sum()
```

```python
# Imputing the missing values of Balance column

df7['Balance']=df7['Balance'].fillna(df7['Balance'].mean())
```

```python
# checking the null value for every column

df7.isnull().sum()
```

```python
#checking the duplicates

df7.loc[df7.duplicated(), :]
```

```python
#removing the duplicates value

df7=df7.drop_duplicates()
```

```python
#Resetting the index

df7.reset_index(drop=True)
df7.sort_index(axis=1, inplace = True)
```

```python
df7
```

```python
#imputing the missing value using interpolation method

df8=df7.interpolate()
```

```python
#checking the null value of every column
df8.isnull().sum()
```

```python
# Changing the datatype of creditScore from float to int

df8["CreditScore"]=df8["CreditScore"].astype(int)
```

```python
df8.head()
```

```python
# Rearranging the final dataframe as suggested in problem statement

df8 =df8[["RowNumber","CustomerId","Surname","CreditScore","Geography","Gender","Age","Tenure","Balance","NumOfProducts","HasChckng","IsActiveMember","EstimatedSalary","Exited"]]
```

```python
df8.head()
```

```python
# Apply lambda function to remove leading/trailing spaces if any

df8 = df8.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df8.head()
```

```python
#how many users are from different geograpghy

df8.groupby('Geography').size()
```

```python
#how many of them have left the PUM services

df8.groupby('Exited').size()[0]
```

```python
#3how many of them exited the company or exists 

df8.groupby('Exited').size()
```

```python
#How many of them have left the PUM services as per the zone.

df8.groupby(['Geography','Exited']).size()
```

```python
#How many of them have left the PUM services as per the Gender.

df8.groupby(['Gender','Exited']).size()
```

```python
# writing the final dataframe into csv if required

df8.to_csv('Customer_final_data.csv', index= False)
```
